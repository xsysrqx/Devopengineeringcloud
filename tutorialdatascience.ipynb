```python
# Importing pandas library
import pandas as pd

# Creating a data frame df.
df = pd.DataFrame({'Date':['11/05/19', '12/05/19','19/05/19',
'25/05/19', '26/05/19', '1/06/19'], 'Time Worked': [3, 3, 4, 3, 3,
4], 'Money Earned': [33.94, 33.94, 46, 33.94, 33.94, 46]})

# Head display only the top 5 rows from the data frame.
df.head()
```

```python
# Adding more rows
df2 = pd.DataFrame({'Date': ['10/06/19', '12/06/19', '14/06/19'],
                  'Tme Worked': [3, 4, 3],
                  'Money Earned': [33.94, 46, 33.94]})
df2
```

```python
# Appending the rows of the old data frame to the new data frame.

df = df.append(df2, ignore_index=True, sort = False)
df.head()
```

```python
Total_earnings = df['Money Earned'].sum()
Total_time = df['Time Worked'].sum()

print("You have earned total of ====>" ,round(Total_earnings), "CAD")
print("-------------------------------------------------------")
print("You have worked for a total of ====>", Total_time, "hours")
```

```python
# Plotting a bar graph using pandas library.
df.plot(x ='Date', y='Money Earned', kind = 'bar')
```

```python
# Function payroll contains calculation of total money.
def payroll():

  name = input("Enter the name of the employee: ==> ")
  hours = int(input("Enter the hours worked by the employee ==>"))
  rate = float(input("Enter the pay rate for one hour ==> "))

  total_money = hours * rate

  print("The total money earned by ", name, "for working ", hours,
        "hours is", round(total_money), "CAD")

  payroll()

```
